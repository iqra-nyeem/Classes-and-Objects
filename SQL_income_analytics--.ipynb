# Filter in SQL - Using WHERE clause

```python
# Get all records of gender 'Male'
query = 'select * from income_train where sex = "Male"'
pd.read_sql_query(query,conn)
```

```python
#SELECT a single column
```

```python
# print a single value
query = 'select Workclass from income_train'
pd.read_sql_query(query,conn)
```

```python
# UNIQUES values DISTINCT
```

```python
query = 'select distinct(Workclass) from income_train'
pd.read_sql_query(query,conn)
```

```python
# SUM in SQL
```

```python
# print a single value of hours per week 
query = 'select "Hours.Per.Week" from income_train where sex = "Male"'
pd.read_sql_query(query,conn)
```

```python
# Gathering All Males for a week would accumalate how many hours of work
query = 'select SUM("Hours.Per.Week") from income_train where Sex = "Male"'
pd.read_sql_query(query,conn)
```

# GROUP BY

```python
#Gathering all Male and Female for a week would accumalate how many hours of work
query = 'select Sex, SUM("Hours.Per.Week") from income_train GROUP BY Sex'
pd.read_sql_query(query,conn)
```

```python
query = 'select Sex, count(*) from income_train GROUP BY Sex'
pd.read_sql_query(query,conn)
```

# ORDER BY 

```python
query = 'select * from income_train ORDER BY "Hours.Per.Week"'
pd.read_sql_query(query,conn)
```

```python
query = 'select * from income_train ORDER BY "Hours.Per.Week" DESC'
pd.read_sql_query(query,conn)
```

```python
query = 'select * from income_train ORDER BY "Hours.Per.Week","Age"DESC'
pd.read_sql_query(query,conn)
```

```python
query = 'select * from income_train ORDER BY "Hours.Per.Week"DESC , "Age"DESC'
pd.read_sql_query(query,conn)
```
