```python
class Lepidoptera:
    def __init__(self,Class,Scientific_Name,Classification,Kingdom):
       
        self.Class = Class
        self.Scientific_Name = Scientific_Name
        self.Classification = Classification
        self.Kingdom = Kingdom
    def LifeSpan(self):
        return "Its Lifespan is '12 Months'."
    
class Papilionoiden(Lepidoptera):      
    def Eggs(self):
         return "The Papilionoiden's 'Eggs' Size is '1.5mm' \n Its Color is greenish-White"
    def Larvae(self):
        return "The Papilionoiden's 'Larvae' Size is '2.4 inc' \n Its Color is green"
    def Pupae(self):
        return "The Papilionoiden's 'Pupae' Size is '2 cm' \n Its Color is Green or Brown"    
    
    def Butterfly(self):
        return "It is the Largest Butterfly in the World \nIt is Dark Brown with 'Yellow' Sports and Bands"

class Pieridae(Papilionoiden):
    def number_of_Species(self):
        return 'Pieridae family has about 76 genera and 1,100 species.'
class Lycaenidae(Papilionoiden):
    def number_of_Species(self):
        return 'Lycaenidae family has about 6,000 species worldwide'


class Hesperioidea(Lepidoptera):
    def Eggs(self):
         return "The Hesperioidea's 'Eggs' Size is '1.77mm' \n Its Color is greenish-White"
    def Larvae(self):
        return "The Hesperioidea's 'Larvae' Size is '2.8 inc' \n Its Color is green"
    def Pupae(self):
        return "The Hesperioidea's 'Pupae' Size is '2.5 cm' \n Its Color is Green or Brown"    
    
    def Butterfly(self):
        return "Its colors varies with Sports and Bands"


print("==============Lepidoptera===========")   
A = Lepidoptera('Insecta','Lepidoptera','Amphiesmenoptera','Animalia')
print('Lepidoptera belong to class {} '.format(A.Class))
print('Lepidoptera belong to Kingdom {} '.format(A.Kingdom))
print('Classification of Lepidoptera are {} '.format(A.Classification))
print("Lepidoptera' Scientific_Name is {} ".format(A.Scientific_Name))
print(A.LifeSpan())
print("============Papilionoiden===========")  
B = Papilionoiden('Insecta','Papilionoidea','Amphiesmenoptera','Animalia')
print('Papilionoiden belong to class {} '.format(B.Class))
print('Papilionoiden belong to Kingdom {} '.format(B.Kingdom))
print('Papilionoiden of Butterflies are {} '.format(B.Classification))
print("Papilionoiden' Scientific_Name is {} ".format(B.Scientific_Name))
print(B.Eggs())
print(B.Larvae())
print(B.Pupae())
print(B.Butterfly())
print("=============Hesperioidea===========")
C = Hesperioidea('Insecta','Hesperiidae','Amphiesmenoptera','Animalia')
print('Hesperioidea belong to class {} '.format(C.Class))
print('Hesperioidea belong to Kingdom {} '.format(C.Kingdom))
print('Hesperioidea of Butterflies are {} '.format(C.Classification))
print("Hesperioidea' Scientific_Name is {} ".format(C.Scientific_Name))
print(C.Eggs())
print(C.Larvae())
print(C.Pupae())
print(C.Butterfly())
print('=============Pieridae=================')
D = Pieridae('Insecta','Papilionoidea','Amphiesmenoptera','Animalia')
print(D.number_of_Species())
print('=============Lycaenidae================')
E =  Lycaenidae('Insecta','Papilionoidea','Amphiesmenoptera','Animalia')
print(E.number_of_Species())


```
